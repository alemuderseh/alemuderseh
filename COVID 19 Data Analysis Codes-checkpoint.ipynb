```python
#Import relevant libraries 
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#Import data from local drive using read_csv pandas feature
df = pd.read_csv("D:/Public Health Data Science/1. PHDS Main Course Files 2023_24/3. Major Courses/4. Data Science Fundamentals/Group Assignment/COVID-19-geographic-disbtribution-worldwide2.csv")
```

```python
#See the columns name and the first 5 rows
df.head()
```

```python
#change dateRep column into pyton datetime format to use it as an index
df["dateRep"] = pd.to_datetime(df["dateRep"], format='%m/%d/%Y')
df['cases'] = df['cases'].abs()
df['deaths'] = df['deaths'].abs()
```

```python
#Rename variables into more easily understandable variable name
df = df.rename(columns={"countriesAndTerritories": "country", "popData2019":"population", "continentExp": "continent", 
                        "Cumulative_number_for_14_days_of_COVID19_cases_per_100000": "comulative_14"})
```

```python
#visualize scatter matrix of cases, deaths with population
from pandas.plotting import scatter_matrix
attributes = [ "cases", "deaths", "population" ]
scatter_matrix(df[attributes], figsize=(10,5))
```

```python
#Group cases and deaths by country
total_cases_per_country = df.groupby(["country"])["cases"].sum().reset_index()
total_deaths_per_country = df.groupby(["country"])["deaths"].sum().reset_index()
```

```python
#Top 25 cases by country
top_25_cases = total_cases_per_country.sort_values(by=["cases"], ascending=False).head(25)
```

```python
plot = sns.barplot(y="country", x="cases", data=top_25_cases)
plt.xlabel("Number of Cases in millions")
for i, (value,name) in enumerate(zip(top_25_cases["cases"], top_25_cases["country"])):
    plot.text(value,i,f'{value:,.0f}',size=8)
```

```python
#Top 25 deaths by country
top_25_deaths = total_deaths_per_country.sort_values(by=["deaths"], ascending=False).head(25)
```

```python
plot = sns.barplot(y="country", x="deaths", data=top_25_deaths)
for i, (value,name) in enumerate(zip(top_25_deaths["deaths"], top_25_deaths["country"])):
    plot.text(value,i,f'{value:,.0f}',size=7)
```

```python
#Group cases and deaths by continent
total_cases_per_cont = df.groupby(["continent"])["cases"].sum().reset_index()
total_deaths_per_cont = df.groupby(["continent"])["deaths"].sum().reset_index()
```

```python
#Plotting cases by continent
# Sort the DataFrame by 'cases' in descending order
sorted_df = total_cases_per_cont.sort_values(by='cases', ascending=False)

# Create the barplot with the sorted data
sns.barplot(x="continent", y="cases", data=sorted_df)

plt.title("Total Cases in millions by Continent")
plt.xlabel("Continent")
plt.ylabel("Total Cases in millions")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python

#Plotting deaths by continent
sorted_df = total_deaths_per_cont.sort_values(by='deaths', ascending=False)

# Create the barplot with the sorted data
sns.barplot(x="continent", y="deaths", data=sorted_df)

plt.title("Total Deaths by Continent (Descending)")
plt.xlabel("Continent")
plt.ylabel("Total Deaths")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
#Global trend of cases and deaths together
daily_cases = df.groupby(['dateRep', 'country'])['cases'].sum()
daily_deaths = df.groupby(['dateRep', 'country'])['deaths'].sum()
df['cases_per_100k'] = df['cases'] / (df['population'] / 100000)
df['deaths_per_100k'] = df['deaths'] / (df['population'] / 100000)
chosen_country = 'Canada'
country_data = df[df['country'] == chosen_country]
plt.figure(figsize=(10, 6))
plt.plot(country_data['dateRep'], country_data['cases_per_100k'], label='Cases per 100,000')
plt.plot(country_data['dateRep'], country_data['deaths_per_100k'], label='Deaths per 100,000')
plt.xlabel('Date')
plt.ylabel('Rate per 100,000 Population')
plt.title(f'COVID-19 Cases and Deaths per 100,000 in {chosen_country}')
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
total_cases = df['cases'].sum()
total_deaths = df['deaths'].sum()
total_population = df['population'].sum()
total_cases_per_100k = total_cases / (total_population / 100000)
total_deaths_per_100k = total_deaths / (total_population / 100000)
results = {
  'Total Cases': total_cases,
  'Total Deaths': total_deaths,
  'Cases per 100,000 Population': total_cases_per_100k,
  'Deaths per 100,000 Population': total_deaths_per_100k
}
print("Total COVID-19 Cases and Deaths (per 100,000 Population):")
df_results = pd.DataFrame(results, index=[0])
print(df_results.to_string(index=False))
```
